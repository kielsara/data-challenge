```python
import numpy as np
import mne
import matplotlib.pyplot as plt

# Define electrode names including M1 and M2
electrode_names = [
    'Fp1', 'Fpz', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8',
    'FC5', 'FC1', 'FC2', 'FC6', 'T7', 'C3', 'Cz', 'C4',
    'T8', 'CP5', 'CP1', 'CP2', 'CP6', 'P7', 'P3', 'Pz',
    'P4', 'P8', 'POz', 'O1', 'Oz', 'O2', 'M1', 'M2' 
]

# Create an MNE info object for the electrodes
sfreq = 256  # Sampling frequency
info = mne.create_info(ch_names=electrode_names, sfreq=sfreq, ch_types='eeg')

# Calculate the average ERP for the 400 ms +/- 25 ms time window
average_erp_per_electrode = data_400_ms.mean(axis=0)

# Create data for the mastoid electrodes
average_erp_full = np.zeros(32)  # Initialize for 32 channels
average_erp_full[:30] = average_erp_per_electrode  # Fill in the first 30 with existing data
average_erp_full[30] = np.mean(average_erp_per_electrode)  # Mean value for M1
average_erp_full[31] = np.mean(average_erp_per_electrode)  # Mean value for M2

# Define the montage
montage = mne.channels.make_standard_montage('standard_1020')
info.set_montage(montage)  # Set the montage

# Create an EvokedArray for plotting
evoked = mne.EvokedArray(data=average_erp_full[np.newaxis, :].T, info=info)  # Shape: (32, 1)

# Plot the topomap
evoked.plot_topomap(times=[0], size=2.5, show=True)
plt.show()
```

### Graphing Grand Average ERP for All Electrodes for Spanish-English and English-Spanish Translations of Participant 1

```python
import os
import glob
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def process_and_plot_erp(directory_path):
    # Get all relevant files in the directory
    file_pattern = os.path.join(directory_path, '*_english-spanish_translation_1.csv')
    file_pattern2 = os.path.join(directory_path, '*_spanish-english_translation_1.csv')
    file_pattern3 = os.path.join(directory_path, '*_spanish-spanish_translation_1.csv')

    # Combine the two patterns to find all files
    all_files = glob.glob(file_pattern) + glob.glob(file_pattern2) + glob.glob(file_pattern3)

    # List to hold the ERP data from each file
    all_erp_data = []

    for file_path in all_files:
        
        # Load data
        sample_data = pd.read_csv(file_path)
        erp_data = sample_data.values
        
        # Define baseline time points
        baseline_start = 0
        baseline_end = 26 
        
        # Calculate the baseline mean across the baseline window
        baseline_mean = np.mean(erp_data[baseline_start:baseline_end, :], axis=0)
        
        # Subtract the baseline mean from each time point
        erp_corrected = erp_data - baseline_mean

        # Calculate indices for 375 ms to 425 ms based on the sampling rate
        sampling_rate = 256  # Hz
        start_index = int(375 / 1000 * sampling_rate)
        end_index = int(425 / 1000 * sampling_rate) + 1
        
        # Extract the time window and compute the average ERP across electrodes for this file
        windowed_data = erp_corrected[start_index:end_index, :]
        average_erp_across_electrodes = windowed_data.mean(axis=1)
        
        # Append the average ERP for this file to the list
        all_erp_data.append(average_erp_across_electrodes)

        # Extract participant number using regex
        participant_number_match = re.search(r'_(\d+)\.csv$', file_path)
        if participant_number_match:
            participant_number = participant_number_match.group(1)
        else:
            participant_number = 'Unknown'

    # Calculate the grand average ERP across all files
    grand_average_erp = np.mean(all_erp_data, axis=0)
    
    # Create time points specifically from 375 ms to 425 ms
    time_points = np.linspace(375, 425, num=len(grand_average_erp))

    # Plotting the grand average ERP for the specific time window
    plt.figure(figsize=(10, 5))
    plt.plot(time_points, grand_average_erp, label='Grand Average ERP', color='blue')
    plt.axvline(x=400, color='red', linestyle='--', label='400 ms (N400)')
    
    plt.title(f'Grand Average ERP Across All Files for Participant {participant_number}')
    plt.xlabel('Time (ms)')
    plt.ylabel('Amplitude (µV)')
    plt.xlim(375, 425)
    plt.legend()
    plt.grid()
    plt.show()

process_and_plot_erp("C:\\Users\\Sara Kiel\\Downloads\\Individual_ERPs\\EEG_Measurements")
```

### Graphing Grand Average ERP for All Electrodes for Spanish-English and English-Spanish Translations of All Participants

```python
import re
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

def process_and_plot_erp_by_participant(directory_path):
    # Get all relevant files in the directory
    file_pattern = os.path.join(directory_path, '*_english-spanish_translation_*.csv')
    file_pattern2 = os.path.join(directory_path, '*_spanish-english_translation_*.csv')
    file_pattern3 = os.path.join(directory_path, '*_spanish-spanish_translation_*.csv')

    # Combine the two patterns to find all files
    all_files = glob.glob(file_pattern) + glob.glob(file_pattern2) + glob.glob(file_pattern3)

    # Dictionary to hold ERP data for each participant
    participant_erp_data = defaultdict(list)

    # Process each file and group by participant number
    for file_path in all_files:
        
        # Extract participant number using regex
        participant_number_match = re.search(r'_(\d+)\.csv$', file_path)
        if participant_number_match:
            participant_number = participant_number_match.group(1)
        else:
            participant_number = 'Unknown'
        
        # Load data
        sample_data = pd.read_csv(file_path)
        erp_data = sample_data.values
        
        # Define baseline time points
        baseline_start = 0
        baseline_end = 26 
        
        # Calculate the baseline mean across the baseline window
        baseline_mean = np.mean(erp_data[baseline_start:baseline_end, :], axis=0)
        
        # Subtract the baseline mean from each time point
        erp_corrected = erp_data - baseline_mean

        # Calculate indices for 375 ms to 425 ms based on the sampling rate
        sampling_rate = 256  # Hz
        start_index = int(375 / 1000 * sampling_rate)
        end_index = int(425 / 1000 * sampling_rate) + 1
        
        # Extract the time window and compute the average ERP across electrodes for this file
        windowed_data = erp_corrected[start_index:end_index, :]
        average_erp_across_electrodes = windowed_data.mean(axis=1)
        
        # Append the average ERP for this file to the participant's list
        participant_erp_data[participant_number].append(average_erp_across_electrodes)

    # Plot the grand average ERP for each participant
    for participant_number, erp_data_list in participant_erp_data.items():
        # Calculate the grand average ERP for this participant
        grand_average_erp = np.mean(erp_data_list, axis=0)
        
        # Create time points specifically from 375 ms to 425 ms
        time_points = np.linspace(375, 425, num=len(grand_average_erp))

        # Plotting the grand average ERP for this participant
        plt.figure(figsize=(10, 5))
        plt.plot(time_points, grand_average_erp, label='Grand Average ERP', color='blue')
        plt.axvline(x=400, color='red', linestyle='--', label='400 ms (N400)')
        
        plt.title(f'Grand Average ERP Across All Files for Participant {participant_number}')
        plt.xlabel('Time (ms)')
        plt.ylabel('Amplitude (µV)')
        plt.xlim(375, 425)
        plt.legend()
        plt.grid()
        plt.show()

process_and_plot_erp_by_participant("C:\\Users\\Sara Kiel\\Downloads\\Individual_ERPs\\EEG_Measurements")
```

### Graphing Grand Average ERP for All Electrodes of Spanish and Non-Spanish Speakers

```python
import re

def process_and_plot_erp(directory_path, language_file_path):
    # Load the language proficiency data
    language_data = pd.read_csv(language_file_path)
    
    # Rename the "participant" column in the language_data to match what main data will use
    language_data.rename(columns={"participant": "participant_number"}, inplace=True)

    # Get all relevant files in the directory
    file_pattern = os.path.join(directory_path, '*_english-spanish_translation_*.csv')
    file_pattern2 = os.path.join(directory_path, '*_spanish-english_translation_*.csv')
    file_pattern3 = os.path.join(directory_path, '*_spanish-spanish_translation_*.csv')

    # Combine the two patterns to find all files
    all_files = glob.glob(file_pattern) + glob.glob(file_pattern2) + glob.glob(file_pattern3)

    # Dictionary to store ERP data for Spanish and Non-Spanish Speakers
    erp_data_by_language = {"Spanish": [], "Non-Spanish": []}

    for file_path in all_files:
        
        # Load data
        sample_data = pd.read_csv(file_path)
        erp_data = sample_data.values
        
        # Define baseline time points
        baseline_start = 0
        baseline_end = 26 
        
        # Calculate the baseline mean across the baseline window
        baseline_mean = np.mean(erp_data[baseline_start:baseline_end, :], axis=0)
        
        # Subtract the baseline mean from each time point
        erp_corrected = erp_data - baseline_mean

        # Calculate indices for 375 ms to 425 ms based on the sampling rate
        sampling_rate = 256  # Hz
        start_index = int(375 / 1000 * sampling_rate)
        end_index = int(425 / 1000 * sampling_rate) + 1
        
        # Extract the time window and compute the average ERP across electrodes for this file
        windowed_data = erp_corrected[start_index:end_index, :]
        average_erp_across_electrodes = windowed_data.mean(axis=1)
        
        # Extract participant number using regex
        participant_number_match = re.search(r'_(\d+)\.csv$', file_path)
        if participant_number_match:
            participant_number = int(participant_number_match.group(1))
        else:
            participant_number = None
        
        # Determine if the participant is a Spanish speaker
        is_spanish_speaker = language_data.loc[language_data["participant_number"] == participant_number, "spanish"].values[0]
        
        # Append ERP data to the appropriate group
        if is_spanish_speaker == 1:
            erp_data_by_language["Spanish"].append(average_erp_across_electrodes)
        else:
            erp_data_by_language["Non-Spanish"].append(average_erp_across_electrodes)

    # Calculate the grand average ERP for each group
    spanish_avg_erp = np.mean(erp_data_by_language["Spanish"], axis=0)
    non_spanish_avg_erp = np.mean(erp_data_by_language["Non-Spanish"], axis=0)
    
    # Create time points specifically from 375 ms to 425 ms
    time_points = np.linspace(375, 425, num=len(spanish_avg_erp))

    # Plotting the grand average ERP for Spanish and Non-Spanish speakers
    plt.figure(figsize=(10, 5))
    plt.plot(time_points, spanish_avg_erp, label='Spanish Speakers', color='blue')
    plt.plot(time_points, non_spanish_avg_erp, label='Non-Spanish Speakers', color='green')
    plt.axvline(x=400, color='red', linestyle='--', label='400 ms (N400)')
    
    plt.title('Grand Average ERP for Spanish and Non-Spanish Speakers (375–425 ms)')
    plt.xlabel('Time (ms)')
    plt.ylabel('Amplitude (µV)')
    plt.xlim(375, 425)
    plt.legend()
    plt.grid()
    plt.show()

process_and_plot_erp("C:\\Users\\Sara Kiel\\Downloads\\Individual_ERPs\\EEG_Measurements", "C:\\Users\\Sara Kiel\\Downloads\\metadata.csv")
```

### 1D Convolutional Neural Network - Just N400

```python
import pandas as pd
import glob
import os
import numpy as np
from sklearn.metrics import classification_report, accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Flatten, Dense, Dropout
from sklearn.metrics import confusion_matrix
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Load metadata
metadata_path = "C:\\Users\\Sara Kiel\\Downloads\\metadata.csv"
metadata = pd.read_csv(metadata_path)
print(metadata.head())
```

```python
# Create a dictionary for participant labels
labels_dict = dict(zip(metadata['participant'], metadata['spanish']))

# Initialize lists to store ERP data and labels
erp_data = []
labels = []

# Load ERP files related to Spanish Bilingualism only
for file in glob.glob("C:\\Users\\Sara Kiel\\Downloads\\Individual_ERPs\\EEG_Measurements\\*_english-spanish_translation_*.csv") + \
             glob.glob("C:\\Users\\Sara Kiel\\Downloads\\Individual_ERPs\\EEG_Measurements\\*_spanish-english_translation_*.csv") + \
             glob.glob("C:\\Users\\Sara Kiel\\Downloads\\Individual_ERPs\\EEG_Measurements\\*_spanish-spanish_translation_*.csv"):
   
    # Extract participant number from filename and remove .csv
    participant_number = os.path.basename(file).split('_')[-1].replace('.csv', '')
    participant_number = int(participant_number)  # Convert to integer

    erp_df = pd.read_csv(file)  # Load the ERP data

    # Append the ERP data
    erp_data.append(erp_df.values)  # Convert to NumPy array

    # Append the corresponding label
    label = labels_dict.get(participant_number)
    if label is not None:
        labels.append(label)
    else:
        print(f"No label found for participant {participant_number}")
```

```python
# Convert lists to NumPy arrays
erp_data = np.array(erp_data)  
labels = np.array(labels)

# Baseline Correction Pre-Stimulus
baseline_start = 0  # corresponds to -100 ms
baseline_end = 26   # approximately 0 ms

# Calculate the baseline mean across the baseline window for each sample and electrode
baseline_mean = np.mean(erp_data[:, baseline_start:baseline_end, :], axis=1, keepdims=True)

# Subtract the baseline mean from each time point to baseline-correct the data
erp_corrected = erp_data - baseline_mean

# Set indices for the time window (375 ms to 425 ms)
start_index = 96  # Represents 375 ms
end_index = 108   # Represents 425 ms

# Slice and average over the selected time window
erp_window_avg = np.mean(erp_corrected[:, start_index:end_index, :], axis=1)

# Reshape for CNN input (n_samples, 30, 1)
erp_window_avg = erp_window_avg.reshape(erp_window_avg.shape[0], 30, 1)  # Each sample has 30 electrodes, reshaped for Conv1D
```

```python
# Define the 1D CNN model
model = Sequential()
model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(30, 1)))  # Input shape is (30, 1)
model.add(Flatten())
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.0001)))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))  # Sigmoid for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Early Stopping (if needed)
# early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model and capture the history
history = model.fit(erp_window_avg, labels, epochs=10, batch_size=32, validation_split=0.2)

# Plot the training and validation accuracy and loss
plt.figure(figsize=(12, 5))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')
plt.legend()

# Show the plot
plt.show()

# Make predictions
predictions = model.predict(erp_window_avg)
predictions = (predictions > 0.5).astype(int).flatten()  # Convert probabilities to binary class labels

# Calculate accuracy
accuracy = accuracy_score(labels, predictions)
print(f'Accuracy: {accuracy:f}')  # Print accuracy percentage

# Generate the classification report
report = classification_report(labels, predictions, target_names=['Non-Spanish Speaking (0)', 'Spanish Speaking (1)'], output_dict=True)
print(classification_report(labels, predictions, target_names=['Non-Spanish Speaking (0)', 'Spanish Speaking (1)']))

# Print the confusion matrix
conf_matrix = confusion_matrix(labels, predictions)
print(conf_matrix)
```
